```python
import pandas as pd #Pandas 모듈 임포트 

CCTV_Seoul = pd.read_csv('../../data/01. CCTV_in_Seoul.csv', encoding='utf-8')
CCTV_Seoul.head(10)
```

```python
CCTV_Seoul.columns #컬럼 확인 하기 
```

```python
CCTV_Seoul.columns[0]
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0] : '구별'}, inplace = True)
CCTV_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('../../data/01. population_in_Seoul.xls', encodling='utf-8')
pop_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('../../data/01. population_in_Seoul.xls',
                            header = 2,
                            parse_cols = 'B, D, G, J, N',
                            encoding = 'utf-8')
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={ pop_Seoul.columns[0] : '구별',
                           pop_Seoul.columns[1] : '인구수',
                           pop_Seoul.columns[2] : '한국인',
                           pop_Seoul.columns[3] : '외국인',
                           pop_Seoul.columns[4] : '고령자'},
                 inplace = True)
pop_Seoul.head()
```

```python
CCTV_Seoul.head()
```

```python
CCTV_Seoul.sort_values(by='소계', ascending=True).head()
```

```python
CCTV_Seoul.sort_values(by='소계', ascending=False).head()
```

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2014년']) / CCTV_Seoul['2013년도 이전'] * 100 
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head()
```

```python
pop_Seoul.drop([0],inplace=True)     #실제 데이터프레임에 영향을 주는 것이 inplace(commit역할)
pop_Seoul.head()
```

```python
pop_Seoul['구별'].unique()
```

```python
pop_Seoul[pop_Seoul['구별'].isnull()]
```

```python
pop_Seoul.drop([26], inplace = True)
pop_Seoul.head()
```

```python
pop_Seoul['구별'].unique()
```

```python
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul.sort_values(by='외국인비율', ascending=False)
```

```python
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on='구별')  #on='구별'은 구별을 중심으로 합친다
data_result.head()
```

```python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']

data_result.head()
```

```python
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
fp1=np.polyfit(data_result['인구수'],data_result['소계'],1)
fp1
```

```python
fl= np.poly1d(fp1)
fx= np.linspace(100000,700000,100)

plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'],data_result['소계'], s=50)
plt.plot(fx,fl(fx),ls='dashed',lw=3, color='g')
plt.xlabel = '인구수'
plt.ylabel = 'CCTV'
plt.grid()
plt.show()
```
