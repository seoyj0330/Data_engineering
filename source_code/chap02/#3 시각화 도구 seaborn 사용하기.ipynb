```python
import matplotlib.pyplot as plt
%matplotlib inline

#seaborn을 사용하기 위해서는 반드시 matpoltlib부터 import할 것

import seaborn as sns
```

```python
import numpy as np
```

```python
x = np.linspace(0, 14, 100)

y1 = np.sin(x) 
y2 = 2*np.sin(x+0.5) 
y3 = 3*np.sin(x+1.0) 
y4 = 4*np.sin(x+1.5)

sns.set_style('whitegrid')

plt.figure(figsize=(10,6)) 
plt.plot(x, y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style('whitegrid')

tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips) #seaborn 사용
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set3') # 흡연자(smoker)로 추가 구분하기 
plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, size=7)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, palette='Set1', size=7) #흡연자 데이터로 비교하기 
plt.show()
```

```python
flights = sns.load_dataset("flights") 
flights.head()
```

```python
flights = flights.pivot("month","year","passengers")
flights.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```

```python
sns.set(style='ticks')
iris = sns.load_dataset('iris')
iris.head(10)
```

```python
sns.pairplot(iris, hue='species') # 꽃 종류별 데이터 비교 분석 하기 
plt.show()
```
