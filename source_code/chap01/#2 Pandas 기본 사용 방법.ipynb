```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']},
                    index =[0,1,2,3])
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']},
                    index =[4,5,6,7])
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11'],
                    'D':['D8','D9','D10','D11']},
                    index =[8,9,10,11])
```

```python
df1
```

```python
result = pd.concat([df1,df2,df3], keys=['x','y','z'])
result
```

```python
result.index
```

```python
result.index.get_level_values(0)
```

```python
df4=pd.DataFrame({'B' : ['B2','B3','B6','B6'],
                  'D' : ['D2','D3','D6','D6'],
                  'F' : ['F2','F3','F6','F6']},
                index=[2,3,6,7])
df4
```

```python
result = pd.concat([df1,df4], axis=1)  #axis가 1이면 가로방향,0이면 세로방향으로 합치기 <일치하는 곳만 데이터가 들어가고 나머지는 nan처리>
result
```

```python
result = pd.concat([df1,df4], axis=1, join='inner') #join='inner'일치하는 것만 합쳐주고 싶을 때
result
```

```python
result = pd.concat([df1,df4], axis=1, join_axes=[df1.index])  #기준이 df1의 인덱스로 df4가 갖고있지않으면 NaN으로 표시
result
```

```python
result = pd.concat([df1,df4], ignore_index = True)   #ignore_index: 인덱스를 무시하고 컬럼에 맞추어서 합치고 싶을때 사용
result
```

```python
left = pd.DataFrame({'key' : ['K0','K4','K2','K3'],
                     'A' : ['A0','A1','A2','A3'],
                     'B' : ['B0','B1','B2','B3']})

right = pd.DataFrame({'key' : ['K0','K1','K2','K3'],
                     'C' : ['C0','C1','C2','C3'],
                     'D' : ['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
result = pd.merge(left,right, on='key')  # on='key' :key라는 컬럼을 중심으로 합치겠다.
result
```

```python
result = pd.merge(left,right, how='left', on='key')
result
```

```python
result = pd.merge(left, right, how='inner', on='key')
result
```

```python
result = pd.merge(left, right, how='outer', on='key')
result
```
