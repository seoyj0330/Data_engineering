```python
import pandas as pd
```

```python
month_m = pd.read_excel('../../data/month_movie.xls', encoding='utf-8')
month_m.head()
```

```python
month_p = pd.read_excel('../../data/person_movie.xls', encoding='utf-8')
month_p.head()
```

```python
month_m['1월비율'] = month_m['1월']/month_m['소계']*100
month_m['2월비율'] = month_m['2월']/month_m['소계']*100
month_m['3월비율'] = month_m['3월']/month_m['소계']*100
month_m.head()
```

```python
merge_m = pd.merge(month_m,month_p,on='영화등급')
merge_m
```

```python
merge_m.set_index('영화등급', inplace=True)
merge_m.head()
```
